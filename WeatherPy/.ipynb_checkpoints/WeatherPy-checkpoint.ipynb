# WeatherPy

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import random
from scipy.stats import linregress

# Import the OpenWeatherMap API key
from api_keys import weather_api_key

# Import citipy to determine the cities based on latitude and longitude
from citipy import citipy
```

### Generate the Cities List by Using the `citipy` Library

### Use the OpenWeatherMap API to retrieve weather data from the cities list generated in the started code

```python
#Initialize variables
cities_tried=[]      #Names of cities requested to API
cities_data=[]       #Store dictionaries with retrieved data from API
cities_found=[]      #Store names of cities from which data was retrieved correctly
cities_target=500    #Amount of cities wanted for analysis
```

```python
#Keep requesting information until having the necessary datapoints (500 default)
while len(cities_found) < cities_target:
    
    #Random latitude and longitude values in ranges
    lat=random.randrange(-90,91)               
    lon=random.randrange(-180,181)              
    
    #Match lat and lon to find nearest city and country code using citipy
    city=citipy.nearest_city(lat, lon).city_name                   
    country=citipy.nearest_city(lat, lon).country_code
    
    #If city hasn't been requested 
    if city not in cities_tried:                
        
        #Add to the city requested list
        cities_tried.append(city)               
        
        #Retrieve json from API using city, country, api_key and specifying units as "metric"
        url=f"https://api.openweathermap.org/data/2.5/weather?q={city},{country}&appid={weather_api_key}&units=metric"
        data=requests.get(url).json()
        
        #Try to store the retreived data in a dictionary inside a list, if successful, append to cities_found.
        try:
            cities_data.append({"lon":data["coord"]["lon"],
                     "lat":data["coord"]["lat"],
                     "max_temp":data["main"]["temp_max"],
                     "humidity":data["main"]["humidity"],
                     "clouds":data["clouds"]["all"],
                     "wind":data["wind"]["speed"],
                     "country":data["sys"]["country"],
                     "date":data["dt"]
                    })
            cities_found.append(city)
            print(f'Found {len(cities_found)} out of {cities_target} ({round(len(cities_found)/cities_target*100)}%) | {city}, {data["sys"]["country"]}')
        
        #If city isn't found in API, throw error message and don't account for city in cities_found.
        except KeyError:
            print(f'{city}:{data["message"]}')
```

---

```python
# Convert the cities weather data into a Pandas DataFrame
cities_df=pd.DataFrame(cities_data,index=cities_found)
cities_df=cities_df.reset_index(drop=False)
cities_df=cities_df.rename(columns={"index":"city"})

# Show Record Count
cities_df.count()
```

```python
# Display sample data
cities_df.head()
```

```python
# Export the City_Data into a csv
cities_df.to_csv("../output_data/cities.csv",index=False)
```

### Create the Scatter Plots Requested

#### Latitude Vs. Temperature

```python
def plot_scatter(df,x_column,y_column):
    plt.figure(figsize=(8,6))
    plt.scatter(x=df[x_column],y=df[y_column], alpha=0.8, edgecolors="black")
    plt.grid(True)
    plt.xlabel(proper_names[x_column])
    plt.ylabel(proper_names[y_column])
    plt.title(f"City {proper_names[x_column]} vs {proper_names[y_column]} ({request_date})")
    plt.savefig(f"../output_data/{x_column}_vs{y_column}.png")
    plt.show()
```

## Requirement 2: Compute Linear Regression for Each Relationship


```python
# Define a function to create Linear Regression plots
def regress_plot(df,x_column,y_column,xcoord,ycoord):
    [slope, intersection, rvalue, pvalue, err]=linregress(df[x_column],df[y_column])
    plt.figure(figsize=(8,6))
    figure1=plt.scatter(x=df[x_column],y=df[y_column], alpha=0.8, edgecolors="black")
    plt.grid(True)
    plt.title(f"{proper_names[x_column]} vs {proper_names[y_column]} ({request_date})")
    plt.xlabel(proper_names[x_column])
    plt.ylabel(proper_names[y_column])
    y_values=slope*df[x_column]+intersection
    plt.plot(df[x_column],y_values, color="r")
    plt.annotate(f"{round(slope,3)}x + {round(intersection,3)}",xy=(xcoord,ycoord),color="r",fontsize=14)
    print(f"The R squared value is: {round(rvalue**2,3)}")
    plt.show()
```

```python
# Create a DataFrame with the Northern Hemisphere data (Latitude >= 0)
northern_hemi_df=cities_df[cities_df["lat"]>=0].copy()
# Display sample data
northern_hemi_df.head()
```

```python
# Create a DataFrame with the Southern Hemisphere data (Latitude < 0)
southern_hemi_df=cities_df[cities_df["lat"]<0].copy()
# Display sample data
southern_hemi_df.head()
```

###  Temperature vs. Latitude Linear Regression Plot

```python
# Linear regression on Northern Hemisphere
regress_plot(northern_hemi_df,"lat","max_temp",0,0)
```

```python
# Linear regression on Southern Hemisphere
regress_plot(southern_hemi_df,"lat","max_temp",-50,30)
```

**Discussion about the linear relationship:** 
* We can observe a stronger linear correlation in the northern hemisphere, where we could predict the max temperature in a city with a high accuracy. 
* In the south we have a weaker correlation, which might be due to not having cities around -80 latitudes the way we do for the northern hemisphere. 
- Both regressions imply that the temperatures get hotter as que get closer to the equator.

### Humidity vs. Latitude Linear Regression Plot

```python
# Northern Hemisphere
regress_plot(northern_hemi_df,"lat","humidity",50,20)
```

```python
# Southern Hemisphere
regress_plot(southern_hemi_df,"lat","humidity",-20,40)
```

**Discussion about the linear relationship:**  
* We can observe a weak linear correlation in the northern hemisphere. 
* There is almost no correlation in the southern hemisphere. 
* Both regressions imply that the humidity increases with the latitude, in a steeper and more linear way after crossing the equator, with less places having low humidity after latitude 50.

### Cloudiness vs. Latitude Linear Regression Plot

```python
# Northern Hemisphere
regress_plot(northern_hemi_df,"lat","clouds",35,43)
```

```python
# Southern Hemisphere
regress_plot(southern_hemi_df,"lat","clouds",-40,80)
```

**Discussion about the linear relationship:** 
* According to the slope on both plots, cloudiness might increase with latitude. 
* However, the data is very spread out in both directions of the regression meaning latitude is not a good linear predictor of cloudiness.

### Wind Speed vs. Latitude Linear Regression Plot

```python
# Northern Hemisphere
regress_plot(northern_hemi_df,"lat","wind",20,12.5)
```

```python
# Southern Hemisphere
regress_plot(southern_hemi_df,"lat","wind",-55,7)
```
